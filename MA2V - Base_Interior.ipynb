```python
from flask import Flask, send_file
import pandas as pd

app = Flask(__name__)

# Carregamento do arquivo Excel para um DataFrame
df = pd.read_excel('/home/rodrigo/Downloads/MA2V - base_final_interior.xlsx')

# Seleção das colunas desejadas
colunas_selecionadas = ['CNPJ', 'RAZAO_SOCIAL', 'NOME_FANTASIA', 'SITUACAO_CADASTRAL', 
                        'CONTATO_VAZIO', 'CORREIO_ELETRONICO', 'DDD_1', 'TELEFONE_1', 'DATA_SITUACAO_CADASTRAL', 'MOTIVO_SITUACAO_CADASTRAL',
                        'DESCRICAO_CNAE', 'TIPO_LOGRADOURO', 'LOGRADOURO', 'NÚMERO', 'COMPLEMENTO', 'BAIRRO', 
                        'CEP', 'UF', 'NOME_MUNICIPIO',  'DDD_2', 'TELEFONE_2', 'DDD_FAX', 'FAX', 'CAPITAL_SOCIAL']

# Divisão do DataFrame em partes de 10.000 linhas
partes = [df[i:i+10000] for i in range(0, len(df), 10000)]

# Adição de uma coluna para enumerar as linhas de forma cumulativa para cada parte
for idx, parte in enumerate(partes):
    parte['linha_cumulativa'] = parte.index + 1  # Começando em 1
    nome_arquivo = f'MA2V - base_final_interior - Parte {idx+1}.xlsx'
    parte[colunas_selecionadas + ['linha_cumulativa']].to_excel(nome_arquivo, index=False)

@app.route('/download/<int:num_parte>')
def download_file(num_parte):
    nome_arquivo = f'MA2V - base_final_interior - Parte {num_parte}.xlsx'
    return send_file(nome_arquivo, as_attachment=True)

if __name__ == '__main__':
    app.run()
```
